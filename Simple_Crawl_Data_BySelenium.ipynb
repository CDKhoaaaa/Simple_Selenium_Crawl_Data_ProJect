```python
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
```

#### *Đọc file txt* ####

```python
with open('ma_so_thi_sinh_binh_thuan.txt', 'r') as file:
    ma_so_thi_sinh = file.read().splitlines()
```

#### *Tạo driver* ####

```python
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
driver.maximize_window()
driver.get('https://vtv.vn/tra-cuu-diem-thi-2023.htm')
time.sleep(2)
```

#### *Nhập dữ liệu và ấn vào nút tìm kiếm trên trang web https://vtv.vn/tra-cuu-diem-thi-2023.htm* ####

```python
result_list = []

for ma_so in ma_so_thi_sinh:
    result_dict = {}
    # Tìm phần tử là ô tìm kiếm
    search_box = driver.find_element(By.ID, "keyworddiemthi")
    search_box.clear()
    search_box.send_keys(str(ma_so))

    # Tìm phần tử là nút search và thực hiện click
    search_button = driver.find_element(By.ID, "btnSearchDiemThi")
    search_button.click()
    time.sleep(0.5)

    # Tìm phần tử là bảng kết quả tìm kiếm
    search_box = driver.find_element(By.ID, "searchresult")
    text = search_box.text
    lst = text.split('\n')

    for item in lst:
        # Tách chuỗi theo dấu ':', lấy phần bên phải là giá trị
        parts = item.split(':')
        key = parts[0].strip()
        value = parts[1].strip()
        result_dict[key] = value
    # Thêm từ điển của thí sinh vào danh sách
    result_list.append(result_dict)

df = pd.DataFrame(result_list)
df.columns = ['SBD', 'Toan', 'Van', 'Ly', 'Hoa', 'Sinh', 'Su', 'Dia', 'GDCD', 'NgoaiNgu', 'MaNgoaiNgu']
print(df)
```

#### *Thêm cột tên ngoại ngữ* ####

```python
language = {
    'N1': 'Tiếng Anh',
    'N2': 'Tiếng Nga',
    'N3': 'Tiếng Pháp',
    'N4': 'Tiếng Trung',
    'N5': 'Tiếng Đức',
    'N6': 'Tiếng Nhật'
}
df['TenNgoaiNgu'] = df['MaNgoaiNgu'].replace(language)
print(df)
```

#### *Thêm cột tên mã cụm thi* ####

```python
df['MaCumThi'] = 47.0
print(df)
```

#### *Thêm cột tên cụm thi* ####

```python
df['TenCumThi'] = 'Sở GDĐT Bình Thuận'
print(df)
```

#### *Thống kê số lượng điểm thi của từng môn* ####

```python
subject_counts = df[['Toan', 'Ly', 'Hoa', 'Sinh', 'Van', 'Su', 'Dia', 'GDCD', 'NgoaiNgu']].apply(pd.to_numeric, errors='coerce').count()
subject_counts = subject_counts.reset_index()
subject_counts = subject_counts.transpose()
subject_counts.columns = subject_counts.iloc[0]
subject_counts = subject_counts[1:]
subject_counts = subject_counts.astype(int)

subject_counts.columns = ['SL_Toan', 'SL_Ly', 'SL_Hoa', 'SL_Sinh', 'SL_Van', 'SL_Su', 'SL_Dia', 'SL_GDCD', 'SL_NgoaiNgu']
subject_counts
```
